# 🤖 Minecraft Bot Helper - Google Colab

Запустите умного Minecraft бота прямо в Google Colab!

## 🚀 Возможности:
- ✅ Умный ИИ чат с игроками
- ✅ Команды для автоматизации
- ✅ Поддержка Minecraft 1.21+
- ✅ Работает 12 часов бесплатно

## 📋 Инструкция:
1. Запустите все ячейки по порядку
2. Введите данные вашего сервера
3. Настройте OpenAI API (опционально)
4. Запустите бота!

```python
# 📦 Установка зависимостей
print("🔧 Устанавливаем Node.js и зависимости...")

!curl -fsSL https://deb.nodesource.com/setup_18.x | sudo -E bash -
!sudo apt-get install -y nodejs

# Клонируем репозиторий
!git clone https://github.com/uhh1fsdfss/MinecraftBots.git
%cd MinecraftBots

# Устанавливаем npm пакеты
!npm install

print("✅ Установка завершена!")
```

```python
# ⚙️ Конфигурация бота
import os
from google.colab import files
import ipywidgets as widgets
from IPython.display import display, HTML

print("⚙️ Настройка бота...")

# Создаем форму для ввода данных
server_host = widgets.Text(
    value='localhost',
    placeholder='Введите IP сервера',
    description='Сервер:',
    disabled=False
)

server_port = widgets.IntText(
    value=25565,
    description='Порт:',
    disabled=False
)

bot_username = widgets.Text(
    value='ColabBot',
    placeholder='Имя бота',
    description='Имя бота:',
    disabled=False
)

openai_key = widgets.Password(
    value='',
    placeholder='Введите OpenAI API ключ (опционально)',
    description='OpenAI ключ:',
    disabled=False
)

minecraft_version = widgets.Dropdown(
    options=['1.21.1', '1.20.4', '1.19.4', '1.18.2'],
    value='1.21.1',
    description='Версия MC:',
    disabled=False
)

print("📝 Заполните данные для подключения:")
display(server_host, server_port, bot_username, openai_key, minecraft_version)

# Сохраняем конфигурацию
def save_config():
    env_content = f"""# Minecraft Server Configuration
MINECRAFT_HOST={server_host.value}
MINECRAFT_PORT={server_port.value}
MINECRAFT_VERSION={minecraft_version.value}

# Bot Configuration
BOT_USERNAME={bot_username.value}
BOT_AUTH=offline

# OpenAI Configuration
OPENAI_API_KEY={openai_key.value or 'your_openai_api_key_here'}
OPENAI_MODEL=gpt-3.5-turbo

# Bot Settings
MAX_CHAT_HISTORY=30
RESPONSE_DELAY=2000
AUTO_RECONNECT=true
LOG_LEVEL=info
"""
    
    with open('.env', 'w') as f:
        f.write(env_content)
    
    print("✅ Конфигурация сохранена!")

# Кнопка для сохранения
save_button = widgets.Button(
    description='💾 Сохранить настройки',
    button_style='success'
)

save_button.on_click(lambda b: save_config())
display(save_button)
```

```python
# 🚀 Запуск бота
import subprocess
import threading
import time
from IPython.display import clear_output

print("🤖 Запускаем Minecraft бота...")
print("📊 Статус: Подключение к серверу...")

# Функция для запуска бота в фоне
def run_bot():
    try:
        process = subprocess.Popen(
            ['npm', 'start'],
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            universal_newlines=True,
            bufsize=1
        )
        
        # Читаем вывод в реальном времени
        for line in process.stdout:
            print(line.strip())
            
    except Exception as e:
        print(f"❌ Ошибка запуска: {e}")

# Запускаем бота в отдельном потоке
bot_thread = threading.Thread(target=run_bot)
bot_thread.daemon = True
bot_thread.start()

print("\n🎮 Бот запущен! Проверьте логи выше.")
print("💡 Совет: Бот будет работать пока работает эта ячейка")
print("⏰ Google Colab: до 12 часов бесплатно")

# Показываем статус
status_html = """
<div style="background: #e8f5e8; padding: 15px; border-radius: 10px; margin: 10px 0;">
    <h3>🤖 Статус бота</h3>
    <p><strong>Сервер:</strong> {host}:{port}</p>
    <p><strong>Имя бота:</strong> {username}</p>
    <p><strong>Версия MC:</strong> {version}</p>
    <p><strong>ИИ чат:</strong> {'✅ Включен' if openai_key.value else '❌ Отключен'}</p>
</div>

<div style="background: #fff3cd; padding: 15px; border-radius: 10px; margin: 10px 0;">
    <h4>📋 Основные команды:</h4>
    <ul>
        <li><code>!help</code> - список всех команд</li>
        <li><code>!come</code> - бот подойдет к вам</li>
        <li><code>!follow &lt;игрок&gt;</code> - следовать за игроком</li>
        <li><code>!mine &lt;блок&gt;</code> - добыть блок</li>
        <li><code>!build tower</code> - построить башню</li>
        <li><code>!status</code> - статус бота</li>
    </ul>
</div>
""".format(
    host=server_host.value,
    port=server_port.value,
    username=bot_username.value,
    version=minecraft_version.value
)

display(HTML(status_html))

# Держим ячейку активной
try:
    while True:
        time.sleep(60)
        print(f"⏰ Бот работает... {time.strftime('%H:%M:%S')}")
except KeyboardInterrupt:
    print("\n🛑 Остановка бота...")
```
